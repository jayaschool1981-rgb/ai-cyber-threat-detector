# CICIDS2017 - Quick EDA
Load a CSV chunk and inspect basic distributions.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

csv_path = Path('data/raw/CICIDS2017_sample.csv')  # update if needed
df = pd.read_csv(csv_path)
df.head()
```

```python
df['Label'].value_counts().plot(kind='bar')
plt.title('Label distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```
